# Monthly Pharmaceutical Sales Report

This notebook provides a structured analysis of monthly pharmaceutical sales data for a Nigerian pharmacy. It includes sales, expenses, and profit for pharmaceutical products.

## Data Import and Setup
We import the data into a pandas DataFrame. The data includes monthly sales and expenses for various pharmaceutical products. Profit is calculated by subtracting expenses from sales.

```python
import pandas as pd

# Pharmaceutical sales data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [50000, 52000, 48000, 56000, 60000],  # in Naira
    'Expenses': [30000, 31000, 29000, 33000, 35000]  # in Naira
}

# Load data into a DataFrame
df = pd.DataFrame(data)

# Calculate Profit
df['Profit'] = df['Sales'] - df['Expenses']
df
```

## Data Summary
This section provides a quick overview of the monthly pharmaceutical sales data, including sales, expenses, and profit.

```python
# Display summary statistics
df.describe()
```

## Data Visualization
The following plot visualizes the monthly pharmaceutical sales, expenses, and profit.

```python
import matplotlib.pyplot as plt

# Plotting the pharmaceutical sales data
plt.figure(figsize=(8, 4))
plt.plot(df['Month'], df['Sales'], label='Sales (Naira)', marker='o', color='blue')
plt.plot(df['Month'], df['Expenses'], label='Expenses (Naira)', marker='o', color='red')
plt.plot(df['Month'], df['Profit'], label='Profit (Naira)', marker='o', color='green')
plt.title('Monthly Pharmaceutical Sales Report')
plt.xlabel('Month')
plt.ylabel('Amount (Naira)')
plt.legend()
plt.grid()
plt.show()
```

# Sales Made In Product Categories

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data setup
np.random.seed(42)
data = {
    "Product_ID": range(1, 21),
    "Product_Name": [
        "Amlodipine", "Chloroquine", "Ibuprofen", "Metformin", "Vitamin C",
        "Paracetamol", "Moisturizer", "Cleanser", "Hand Sanitizer", "Insulin",
        "Aspirin", "Folic Acid", "Anti-fungal Cream", "Anti-septic Soap", "Oral Contraceptive", "Calcium Supplement", "Face Wash", "Antihistamine",
        "Hydrocortisone Cream", "Thyroid Hormone"
    ],
    "Units_Sold": np.random.randint(50, 500, size=20),
    "Unit_Price": np.round(np.random.uniform(5, 50, size=20), 2)
}

# Create DataFrame
df = pd.DataFrame(data)

df['Total_Sales'] = df['Units_Sold'] * df['Unit_Price']
```

```python
# Define category mapping
category_mapping = {
    "Amlodipine": "Antihypertensive",
    "Chloroquine": "Antimalarial",
    "Ibuprofen": "Analgesic",
    "Metformin": "Antidiabetic",
    "Vitamin C": "Supplement",
    "Paracetamol": "Analgesic",
    "Moisturizer": "Cosmetic",
    "Cleanser": "Cosmetic",
    "Hand Sanitizer": "Cosmetic",
    "Insulin": "Antidiabetic",
    "Aspirin": "Analgesic",
    "Folic Acid": "Supplement",
    "Anti-fungal Cream": "Antiseptic",
    "Anti-septic Soap": "Antiseptic",
    "Oral Contraceptive": "Contraceptive",
    "Calcium Supplement": "Supplement",
    "Face Wash": "Cosmetic",
    "Antihistamine": "Antiallergic",
    "Hydrocortisone Cream": "Antiseptic",
    "Thyroid Hormone": "Hormone Replacement"
}

df["Category"] = df["Product_Name"].map(category_mapping)
```

```python
# Calculate total sales per category
category_sales = df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False).reset_index()
category_sales
```

```python
# Plot total sales by category
sns.set_style('whitegrid')

# Sort categories by total sales
category_sales_sorted = category_sales.sort_values(by='Total_Sales', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=category_sales_sorted, x='Category', y='Total_Sales', palette='viridis')
plt.title('Total Sales by Drug Category')
plt.xlabel('Drug Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.show()
```
